```python
lr=LinearRegression()
X=df[['enginesize']]
y=df['price']
lr.fit(X,y)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X,y,color="blue",label="Actual Data Points")
```

```python
y_predicted=lr.predict(X)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(y, y_predicted, alpha=0.6)
plt.title("Actual Price VS Predicted Price")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
# plt.plot(y,y_predicted,'--',color='red',label="Linear Regression")
min_price = min(y)
max_price = max(y)
plt.plot([min_price, max_price], [min_price, max_price], '--', color='red', linewidth=2, label="Perfect Prediction")
plt.legend()
plt.show()
```
