```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
df=pd.read_csv('CarPrice_Assignment.csv')
print(f"No of Rows: {df.shape[0]}")
print(f"No of Columns: {df.shape[1]}")
```

Here I am going to build a linear regression model that can be able to predict the price of a car


```python
lr=LinearRegression()
x=df[['enginesize']]
y=df['price']
lr.fit(x,y)
```

```python
x_input=float(input("Enter the Engine Size: "))
y_predict=lr.predict([[x_input]])
print(y_predict)
```

```python
Intercept=lr.intercept_
print(f"Intercept: {Intercept}")
Slope=lr.coef_
print(f"Slope: {Slope}")
```

As we know that The simple Linear regression model can be represented in the mathematical form as ***y=mx+c*** (*Where* ***y=dependent variable(output/target variable*** , ***x=independent variable*** , ***m=slope(that is coef_) , c=intercept(that is intercept_)***))
